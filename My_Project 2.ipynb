```python
import pandas as pd   #1st and most important
import numpy as np    #2nd and important too
from datetime import datetime   #When we have to change columns with dates
import matplotlib.pyplot as plt  #For plotting 
import seaborn as sns         #For ploting, works together with matplotlib.pyplot


#settings
pd.set_option('display.max_columns', None)      #for max columns output
```

```python
df = pd.read_csv('Accidents0515.csv', sep=",", encoding='utf8')
df.head(15)
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df.info()
```

```python
df['Weather_Conditions'].unique()
```

```python
columns = ['Accident_Index', 'Location_Easting_OSGR', 'Location_Northing_OSGR', 'Police_Force', 'Accident_Severity','Number_of_Vehicles'
          , 'Number_of_Casualties','Date','Day_of_Week','Time','Speed_limit'
          ,'Light_Conditions','Weather_Conditions','Urban_or_Rural_Area', 'Did_Police_Officer_Attend_Scene_of_Accident']

df2 = df[columns]
```

```python
df2.head()
```

```python
Light_Conditions = {
                1: 'Daylight',
                2: 'Darkness - lights lit',
                3: 'Darkness - lights unlit',
                4: 'Darkness - no lighting',
                5: 'Darkness - lighting unknown'
}

Accident_Severity = {
    1: 'Fatal',
    2: 'Serious',
    3: 'Slight',
}


Weather_Conditions = {
    1: 'ice',
    2: 'rainy',
    3: 'hail',
    4: 'snow',
    5: 'dry',
    6: 'hot',
    7: 'drizzle',
    8: 'foggy',
    9: 'windy'
}

```

```python
df2['Light_Conditions'] = df2['Light_Conditions'].map(Light_Conditions)
df2['Accident_Severity'] = df2['Accident_Severity'].map(Accident_Severity)
df2['Weather_Conditions'] = df2['Weather_Conditions'].map(Weather_Conditions)
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.isna().sum()
```

```python
df2.dropna(subset=['Light_Conditions'], inplace=True)
```

```python
df2.isna().sum()
```

```python
df2.dropna(subset=['Weather_Conditions'], inplace=True)
```

```python
df2.dropna(subset=['Time'], inplace=True)
```

```python
df2.head().isna().sum()
```

```python
df_final = df2.copy()
```

```python
df_final.head()
```

```python
Casualties_per_day_of_week = df_final.groupby(by='Day_of_Week').count()
```

```python
Casualties_per_day_of_week
```

```python
plt.figure(figsize=(20,10))
plt.bar(Casualties_per_day_of_week.index, Casualties_per_day_of_week['Number_of_Casualties'])
# plt.gca().invert_yaxis() #reverse y axis with Z
plt.title('Casualties per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Casualties')
plt.xticks(Casualties_per_day_of_week.index, ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.show()
None
```
